**Question 9**

which genre has highest liveness and lowest speechliness

Ans:

image.png

```python
import pandas as pd
import numpy as np

# Get the columns of interest
liveness = df['liveness']
speechiness = df['speechiness']

# Find the genre with the highest liveness
highest_liveness_genre = df['genres'][liveness.idxmax()]

# Find the genre with the lowest speechiness
lowest_speechiness_genre = df['genres'][speechiness.idxmin()]

# Return the results

result = (highest_liveness_genre, lowest_speechiness_genre)
```
